```python
import random
c1=c2=c3=c4=c5=c6=0
for x in range(10000) :
    a = random.randint(1,6)
    if a==1 :
        c1 += 1
    elif a==2 :
        c2 += 1
    elif a==3 :
        c3 += 1
    elif a==4 :
        c4 += 1
    elif a==5 :
        c5 += 1
    elif a==6 :
        c6 += 1
print(c1,c2,c3,c4,c5,c6)
```

```python
a1=[[96,84,80],[96,86,76],[76,95,83],[89,96,69],[90,76,91],[82,66,88],[83,86,79],[85,90,83]]

sc1=[]
sc2=[]
sc3=[]

for x in range(8) :
    sc1.append(a1[x][0])
    sc2.append(a1[x][1])
    sc3.append(a1[x][2])
print(f'국어점수 : {sc1}')
print(f'영어점수 : {sc2}')
print(f'수학점수 : {sc3}')
print(f'국어평균 : {round(sum(sc1)/len(sc1),1)}')
print(f'영어평균 : {round(sum(sc2)/len(sc2),1)}')
print(f'수학평균 : {round(sum(sc3)/len(sc3),1)}')
```

```python
sc1=[]
for score in a1:           ## for score in a1 : score에 a1 항목을 할당, 리스트 항목수만큼 반복
    sc1.append(score[0])   ## for 문으로 score에 a1 리스트의 첫번째부터 순서대로 한번씩 할당
    sc2.append(score[1])   ## score[0] = a1 리스트 내부 리스트의 0번(1번째) 항목
    sc3.append(score[2])
print(sc1)
```

```python
## 튜플과 리스트 차이점 : 리스트는 값의 생성,삭제,수정이 가능하지만 튜플은 값을 바꿀 수 없다.
## 프로그램이 실행되는 동안 값이 바뀌면 안되는 경우 튜플을 사용

t1 = (1,5,3,4,2)
t2 = (6,7,8,9,8)
print(t1[1])                ##인덱싱
print(t1[2:4])              ##슬라이싱
print(t1+t2)                ##더하기
print(t1*3)                 ##곱하기
print(len(t1))              ##길이 구하기
print(t2.index(9))          ##요소의 위치 찾기
print(t2.count(8))          ##요소의 갯수
sort_t1 = sorted(t1)        ##정렬된 새로운 리스트 반환 -> 리스트형태로 반환함
print(sort_t1)
```

```python
dict_voca={}

while True:
    eng = input('eng:')
    if eng =='':
        break
    kor = input('kor:')
    dict_voca[eng] = kor
dict_voca
    
```

```python
score = 0
for x in dict_voca:     ## x : 영어단어 : dict_voca의 키값
    a = input(f'{x}:')  ## 영어 -> 한글
    if a == dict_voca[x]:
        score += 1
print(score,'점','/',len(dict_voca),'점')
```

```python
score = 0
for x in dict_voca.keys():
    a = input(f'{dict_voca[x]}:')  ## 한글 -> 영어 반대로
    if a == x:
        score += 1
print(score,'점')
```
