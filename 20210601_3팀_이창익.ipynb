```python
# 조건 제어 반복 : while 구조
# while (조건문) :  형태
# 조건식의 결과가 True인 동안 계속 반복됨 (무한루프에 빠질 수 있음)
# False가 되면 반복을 빠져나온다.
```

```python
i=1
while i<=5:
    print(i,end=' ')
    i=i+1   ## 루프의 조건식에 사용되는 변수 값은 루프 내에서 변경되어야 한다.
```

```python
# q를 입력할 때까지 반복입력
name = ' '
while name != 'q':    ## name이 q가 아닌 조건이 True(name이 q가 아니면)면 계속 반복
    name = input('이름을 입력하세요 : ')
    print(name)
```

```python
name = ' '
while name != '':    ## 엔터를 치면 반복 끝
    name = input('이름을 입력하세요 : ')
    print(name)
```

```python
# 주문량이 0이거나 총 주문량이 100 이상이 되면 중단
sum = 0
num = ''
while sum<=99 and num!=0:
    num = int(input(f'주문량(<{100-sum}) : '))
    sum += num
print('총 :',sum,'개')
```

```python
# n개의 정수를 입력받아 합계 구하기
n = int(input('몇 개의 숫자를 더할까요? :'))
sum = 0
for i in range(n):
    num = int(input('정수 입력 : '))
    sum += num
print('합계출력 : ',sum)
```

```python
# 짝수의 합계 구하기
n = int(input('몇 개의 숫자를 더할까요? :'))
sum = 0
for i in range(n):
    num = int(input('정수 입력 : '))
    if num % 2 ==0:
        sum += num
print('합계출력 : ',sum)
```

```python
# 반복에서 벗어나기 : break
text = 'I have a dream'
letter = input('찾을 문자 : ')

for i in text:
    print(i,end='')
    if i == letter:  # text 문장에서 letter에 저장된 문자를 찾으면 for문의 반복을 멈춤
        break
```
